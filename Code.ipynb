### **Import and Cleaning**

```python
import pandas as pd
import numpy as np
df = pd.read_csv('properties - properties.csv')
df.head()
```

```python
#list column names to determine importance
for col in df.columns: 
    print(col)
```

```python
#sort by state to make external site usage easier 
#df.sort_values(by=['fips_state'])
#df = df.reset_index(drop=True)
#reorder state as first column
col = df.pop('fips_state')
df.insert(0, col.name, col)
df = df.sort_values(by=['fips_state']).reset_index(drop=True)
df.head()
```

```python
#reorg dataframe with only columns deemed necessary (removing majority null columns as well as physical characteristics not related to ownership)
#removing "owner_std_name2_full" etc was only hesitation point, but any discrepancies in ownership type from there and std1 were negligible
df = df.iloc[:, [0,17,22,42,45,47,51,66,67,73,97]]
```

```python
df.head()
```

```python
#identifying missing data in columns
df.isnull().sum(axis = 0)
```

```python
#replacing null values with empty string
df.fillna('', inplace=True)
```

```python
#combining columns into singular "address" column and dropping previous
all_columns = list(df)
df[all_columns] = df[all_columns].astype(str)

df['address'] = df[['mailing_std_house_nbr', 'mailing_std_street', 'mailing_std_mode','mailing_std_zip5']].agg(' '.join, axis=1)
df['address'] = df['address'].str.rstrip('.0')

df.drop(df.columns[[3, 4, 5, 6,]], axis = 1, inplace = True) 
df.head()
```

```python
#separating those owned by businesses vs individuals, into new dataframe for further analysis 
bizdf = (df[(df.owner_std1_entity_type.str.contains('business entity',case=False))].reset_index(drop=True)).sort_values(by=['fips_state'])

persondf = df[~df.owner_std1_entity_type.str.contains("business entity")]
```

### **Final Dataframe Combination**

```python
biz2 = pd.read_csv('business_owned_fill.csv', index_col=0)
biz2.head()
```

```python
#simplify df columns then rename
biz2 = biz2[['fips_state', 'owner_std1_entity_type','address','Owner Name']]
biz2.columns = ['fips_state','entity_type','address','owner name']
biz2.head()
```

```python
#load in df of person-owned properties and perform same
persondf = persondf[['fips_state','owner_std1_entity_type','address', 'owner_std_name1_full']]
persondf.columns = ['fips_state','entity_type','address','owner name']
persondf.head()
```

```python
owners_df = pd.concat([biz2, persondf], axis=0)
owners_df.head()
```
